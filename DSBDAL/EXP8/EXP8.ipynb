```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# how the price of the ticket for each passenger is distributed using distplot()
sns.histplot(df['fare'])
```

```python
sns.histplot(df['fare'], kde=False)
```

```python
sns.histplot(df['fare'], kde=False, bins=5)
```

```python
sns.displot(df['fare'], kde=True)
```

```python
sns.histplot(df['fare'],bins=30)
```

```python
# 1.
sns.jointplot(x='age', y='fare', data=df)
```

```python
sns.jointplot(x='age', y='fare', data=df, kind='hex')
```
