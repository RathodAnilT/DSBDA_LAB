```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
dataset = sns.load_dataset('titanic')
```

```python
print(dataset)
```

```python
dataset.head()
```

```python
sns.boxplot(x ='sex', y ='age',data = dataset)
```

```python
sns.boxplot(x='sex', y='age',hue='survived',data=dataset)
```
