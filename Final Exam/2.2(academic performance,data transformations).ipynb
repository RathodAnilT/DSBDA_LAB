```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer

# Load the dataset
data = pd.read_csv("academic_performance.csv")

# Step 1: Handle missing values
missing_values = data.isnull().sum()
print("Missing Values:")
print(missing_values)

# Fill missing values with mean for numerical columns
data.fillna(data.mean(), inplace=True)

# Step 2: Check for inconsistencies (unique values of categorical features)
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    print(f"Unique values for {feature}:")
    print(data[feature].unique())
    print()

# Step 3: Apply data transformation (e.g., log transformation) on a numerical variable
# Let's choose 'raisedhands' as an example for transformation
feature_to_transform = 'raisedhands'

# Apply log transformation to 'raisedhands' to reduce skewness
data[feature_to_transform] = np.log1p(data[feature_to_transform])

# Plot histogram of the transformed variable
plt.figure(figsize=(8, 6))
sns.histplot(data[feature_to_transform], kde=True)
plt.title(f"Distribution of {feature_to_transform} (after log transformation)")
plt.xlabel(feature_to_transform)
plt.ylabel("Frequency")
plt.show()

# Optional: Save the cleaned and transformed dataset to a new CSV file
data.to_csv("cleaned_academic_performance.csv", index=False)
```
