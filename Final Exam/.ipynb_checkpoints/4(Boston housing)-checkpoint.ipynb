```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load your dataset (replace this with your dataset loading code)
data = pd.read_csv(r"housing.csv")

# Define features (X) and target variable (y)
X = data[['total_rooms', 'population']]
y = data['median_house_value']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred = model.predict(X_test)

# Visualize the predicted values against the actual values
plt.scatter(y_test, y_pred, c='red')
plt.xlabel("True Price ($100)")
plt.ylabel("Predicted Price ($100)")
plt.title("True Price vs Predicted Price")
plt.show()

# Example prediction using specific features
features = np.array([[6,19]])  # Use the same features as used during training
prediction = model.predict(features)
print(f"Prediction of house price using features {features}: {prediction}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dataset=pd.read_csv(r"housing.csv")
dataset

#sns.heatmap(dataset,corr(),annot=True)

X=dataset[["total_rooms","population"]]
y=dataset["median_house_value"]
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(X_train)
print(X_test)
print(y_train)
print(y_test)

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(X_train,y_train)

y_pred = model.predict(X_test)
y_pred

plt.scatter(y_test,y_pred,c='skyblue')
plt.xlabel("True price$1000")
plt.ylabel("predicted price")
plt.title("true price vs predicted price")
plt.show()

features=np.array([[6,19]])
prediction=model.predict(features)

print(f"Prediction for housing price{features}:{prediction}")
```
