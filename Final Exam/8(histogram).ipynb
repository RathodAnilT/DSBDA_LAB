```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

dataset=pd.read_csv("titanic.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['age']=dataset['age'].fillna(dataset['age'].mean())
```

```python
dataset.isnull().sum()
```

```python
dataset['embarked']=dataset['embarked'].fillna(dataset['embarked'].mode()[0])
```

```python
dataset['deck']=dataset['deck'].fillna(dataset['deck'].mode()[0])
```

```python
dataset['embark_town']=dataset['embark_town'].fillna(dataset['embark_town'].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['fare'],kde=True)
```

```python
sns.distplot(dataset['fare'],kde=False)
```

```python
sns.distplot(dataset['fare'],kde=False,bins=10)
```

```python
sns.histplot(x='fare',data=dataset)
sns.set(rc={'figure.figsize':(5,5)})  #1st:width, 2nd:height
```
