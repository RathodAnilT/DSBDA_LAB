```python
#Handling Missing Values 
df.fillna(value=0, inplace=True) # fill missing values with value=0
df.head(10)
```

```python
#Data formatting and Data Transformation 
df ["Total Doses Administered"] = df[ 'Total Doses Administered' ].astype('int64') 
df["Sessions"] = df['Sessions'].astype('int64') 
df['First Dose Administered'] = df['First Dose Administered' ].astype('int64') 
df['Second Dose Administered'] = df['Second Dose Administered'].astype('int64')  
df['Male (Doses Administered)']= df['Male (Doses Administered)'].astype('int64') 
df["Female (Doses Administered)"]= df['Female (Doses Administered)'].astype('int64') 
df['Transgender (Doses Administered)'] = df['Transgender (Doses Administered)'].astype('int64')
df[' Covaxin (Doses Administered)']= df[" Covaxin (Doses Administered)"].astype('int64') 
df['CoviShield (Doses Administered)'] = df['CoviShield (Doses Administered)'].astype('int64') 
df['Sputnik V (Doses Administered)'] = df['Sputnik V (Doses Administered)'].astype('int64') 
df['AEFI'] = df['AEFI'].astype('int64') 
df['18-44 Years (Doses Administered)'] = df['18-44 Years (Doses Administered)'].astype('int64')
df['45-60 Years (Doses Administered)']= df['45-60 Years (Doses Administered)'].astype('int64')
df['60+ Years (Doses Administered)'] = df['60+ Years (Doses Administered)'].astype('int64')  
df['18-44 Years(Individuals Vaccinated)'] = df['18-44 Years(Individuals Vaccinated)'].astype('int64')
df['45-60 Years(Individuals Vaccinated)'] = df['45-60 Years(Individuals Vaccinated)'].astype('int64') 
df['60+ Years(Individuals Vaccinated)'] = df['60+ Years(Individuals Vaccinated)'].astype('int64')
df['Male(Individuals Vaccinated)'] = df['Male(Individuals Vaccinated)'].astype('int64') 
df['Female(Individuals Vaccinated)'] = df['Female(Individuals Vaccinated)'].astype('int64') 
df['Transgender(Individuals Vaccinated)'] =df['Transgender(Individuals Vaccinated)'].astype('int64')
df['Total Individuals Vaccinated'] = df['Total Individuals Vaccinated'].astype('int64') 
```

```python
df.dtypes
```

```python
date = df['Updated On'].str.split('/', expand=True) 
date
```

```python
df['day'] = date[0] 
df['month'] = date[1] 
df['year'] = date[2] 
df['day'] = pd.to_numeric(df['day']) 
df['month'] = pd.to_numeric(df['month']) 
df['year'] = pd.to_numeric(df['year']) 
df['Updated On'] = pd.to_datetime(df['Updated On']) 
df.rename(columns = {'Updated On' : 'Vaccine_date'}, inplace = True) 
df.head(10)
```

```python
#Number of males vaccinated 
number_of_males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print('The number of males vaccinated are: ', number_of_males_vaccinated)
```

```python
#Number of females vaccinated
number_of_females_vaccinated = df['Female(Individuals Vaccinated)'].sum()
print('The number of females vaccinated are: ', number_of_females_vaccinated)
```

```python
#Number of transgenders vaccinated
number_of_transgenders_vaccinated = df['Transgender(Individuals Vaccinated)'].sum()
print('The number of transgenders vaccinated are: ', number_of_transgenders_vaccinated)
```

```python
#Percentage of people vaccinated in India
plt.pie([number_of_males_vaccinated,number_of_females_vaccinated,number_of_transgenders_vaccinated*20], labels = ['Males','Females','Transgenders'], colors=['blue','orange','yellow'], autopct='%1.1f%%')
plt.legend(loc='lower left', shadow=True,fontsize=11) 
plt.title('Gender-wise Vaccination Distribution')
plt.show()
```

```python
df['State'].unique() 
```

```python
#Excluding the value in States which has India 
new_df = df[df['State'] != 'India'] 
new_df.head()
```

```python
new_df['State'].unique()
```

```python
# Statewise Total Vaccination 
statewise_vaccination = new_df.groupby('State')['Total Individuals Vaccinated'].sum().to_frame() #.toframe() used to create a new frame
statewise_vaccination.head(5) 
```

```python
#Plotting total vaccination done statewise
ax = statewise_vaccination.plot(kind='barh', figsize=(10, 15))
plt.ylabel('Number of people vaccinated') 
plt.xlabel('States in India')
plt.grid(True, linewidth=0.5) 

# Annotating the bars with their values
for index, value in enumerate(statewise_vaccination['Total Individuals Vaccinated']):
    plt.text(value, index, f'{value:,.0f}', va='center', ha='left', fontsize=8, fontweight='bold')
plt.title('People Vaccinated Statewise', loc='left', fontweight='bold')

plt.show()
```

```python
# Remove rows where the state is "India"
df = df[df['State'] != 'India']

doses_statewise = df.groupby('State')[['First Dose Administered', 'Second Dose Administered']].sum()

# Plotting state-wise vaccination for both doses horizontally
plt.figure(figsize=(10, 15))  # Adjust figure size as needed

# Spread out the data vertically
bar_width = 0.35
positions = range(len(doses_statewise))

plt.barh(positions, doses_statewise['First Dose Administered'], height=bar_width, label='First Dose')
plt.barh([p + bar_width for p in positions], doses_statewise['Second Dose Administered'], height=bar_width, label='Second Dose')

plt.title('State-wise Vaccination for First and Second Doses in India', loc='left')
plt.xlabel('Number of individuals vaccinated')
plt.ylabel('States in India')
plt.yticks([p + bar_width / 2 for p in positions], doses_statewise.index)
plt.legend()
plt.grid(True, linewidth=0.5)  # Add gridlines
plt.tight_layout()
plt.show()
```
